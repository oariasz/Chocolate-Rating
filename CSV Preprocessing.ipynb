```python
# Function to replace values from a dataframe column
# Returns: new_dataframe, num_replacements_made
'''
def dfcol_replace(dataframe, column, to_replace, replacement):
    # Make a copy of the DataFrame to avoid modifying the original
    df_copy = dataframe.copy()

    # Fill NaN values with a placeholder value
    df_copy[column] = df_copy[column].fillna("")

    # Replace values in the specified column
    replacements_mask = df_copy[column].str.contains(to_replace)
    df_copy.loc[replacements_mask, column] = df_copy.loc[replacements_mask, column].str.replace(to_replace, replacement)

    # Count the number of replacements made
    num_replacements = replacements_mask.sum()

    return df_copy, num_replacements
'''

def dfcol_replace(dataframe, column, to_replace, replacement):
    # Make a copy of the DataFrame to avoid modifying the original
    df_copy = dataframe.copy()

    # Define a function to replace values in a single cell
    def replace_values(cell_value):
        if pd.notna(cell_value):
            return cell_value.replace(to_replace, replacement)
        else:
            return cell_value

    # Apply the replacement function to the specified column
    df_copy[column] = df_copy[column].apply(replace_values)

    # Count the number of replacements made
    num_replacements = (df_copy[column] != dataframe[column]).sum()

    return df_copy, num_replacements
```
